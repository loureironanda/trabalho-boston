Este conjunto de dados contém informações coletadas pelo Serviço de Recenseamento dos EUA sobre habitação na área da Missa de Boston. Ele foi obtido a partir do arquivo StatLib e tem sido usado extensivamente a literatura para benchmarking de algoritmos. No entanto, essas comparações foram feitas principalmente fora do Delve e são, portanto, um tanto suspeitas. O conjunto de dados é pequeno em tamanho, com apenas 506 casos.


```python
## 9-Gere um X de treino e y de treino, X de teste e y da base (Split Train/Test)

from sklearn.cross_validation import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4,random_state=4)
```

```python
## 11-Usando modelo de regressão linear do sklearn, treine o modelo com o X e y de treino

from sklearn.linear_model import LinearRegression 

linreg = LinearRegression()

linreg.fit(x_train, y_train)
```

```python
## 12-Imprima os atributos de "intercept" e "coefficients" que foram gerados

print(linreg.intercept_)
print(linreg.coef_)
```

```python
## 13-imprima o par "feature names" com os "coefficients"
list(zip(feature_cols, linreg.coef_))
```

```python
## 14-Faça uma previsão usando o conjunto de X de teste
y_pred = linreg.predict(x_test)
```

```python
feature_cols = ['CRIM', 'ZN']

x = df[feature_cols]

y = df.target

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)

linreg.fit(x_train, y_train)

y_pred = linreg.predict(x_test)

print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
